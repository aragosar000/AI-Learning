```python
!mamba install seaborn
import numpy as np
import seaborn as sns


def unit_step_func(x):
    return np.where(x >= 0.5, 1, 0)


class Adaline:

    def __init__(self, learning_rate=0.001, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, x, y):
    n_samples, n_features = x.shape

    # Initialize weights and bias
    self.weights = np.random.randn(n_features) * 0.01
    self.bias = np.random.randn() * 0.01

    # Train for several epochs(one complete pass of the entire training dataset)
    for _ in range(self.n_iters):

        # Look at one passenger at a time
        for idx, x_i in enumerate(x):

            # Calculate the raw score
            linear_output = (
                np.dot(x_i, self.weights)
                + self.bias
            )

            # Compare the raw score to the correct answer
            error = y[idx] - linear_output

            # Update every weight
            self.weights += self.lr * error * x_i

            # Update the bias
            self.bias += self.lr * error


    return self

    def predict(self, x):
        # Calculate raw linear output
        linear_output = np.dot(x, self.weights) + self.bias

        # Convert raw output into class 0 or 1
        predictions = unit_step_func(linear_output)

        return predictions
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

print(titanic.head(4))

# Keep only the columns we need
data = titanic[
    ["survived", "sex", "pclass", "age", "alone"]
].copy()

# Convert sex into numbers
data["sex"] = data["sex"].map({
    "male": 0,
    "female": 1
})

# Convert alone into numbers
data["alone"] = data["alone"].map({
    False: 0,
    True: 1
})

# Remove passengers whose age is missing
data = data.dropna()

# Create input data and correct answers
x = data[
    ["sex", "pclass", "age", "alone"]
].to_numpy(dtype=float)

y = data["survived"].to_numpy(dtype=float)

# Empty list to store every accuracy score
accuracies = []

# Test 10 different shuffled splits
for seed in range(10):
    np.random.seed(seed)

    # Create and shuffle passenger row numbers
    indices = np.arange(len(x))
    np.random.shuffle(indices)

    # Split 80% training and 20% testing
    split_index = int(len(x) * 0.8)

    train_indices = indices[:split_index]
    test_indices = indices[split_index:]

    x_train = x[train_indices]
    x_test = x[test_indices]

    y_train = y[train_indices]
    y_test = y[test_indices]

    # Find the training-set mean and standard deviation
    mean = x_train.mean(axis=0)
    std = x_train.std(axis=0)

    # Prevent division by zero
    std[std == 0] = 1

    # Standardize training and testing data
    x_train = (x_train - mean) / std
    x_test = (x_test - mean) / std

    # Train a new ADALINE model
    model = Adaline(
        learning_rate=0.01,
        n_iters=1000
    )

    model.fit(x_train, y_train)

    # Make predictions
    predictions = model.predict(x_test)

    # Calculate and save accuracy
    accuracy = np.mean(predictions == y_test)
    accuracies.append(accuracy)

    print(
        "Seed:",
        seed,
        "Accuracy:",
        accuracy * 100,
        "%"
    )

# Average of all 10 accuracy scores
average_accuracy = np.mean(accuracies)

print(
    "Average accuracy:",
    average_accuracy * 100,
    "%"
)
```
